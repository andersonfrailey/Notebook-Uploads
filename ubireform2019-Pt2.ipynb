```python
def table_values(data, var):
    """
    Function to return the values for each table
    """
    val = data.groupby('bins', as_index=False).apply(weighted_mean, var)
    return val


def table(data, income_measure, bins=None):
    """
    Function to create a table
    Parameters
    ----------
    data: DataFrame containing data from Tax-Calculator
    income_measure: income used for binning
    bins: number of desired bins
    labels: bin labels
    """

    # add bin labels
    data['bins'] = pd.cut(data[income_measure], bins, include_lowest=True)
    floors = []
    for _bin in np.unique(data['bins']):
        floor = min(data[income_measure][data['bins'] == _bin])
        floors.append(floor)
    decile_floor = pd.Series(floors)

    # find values for table
    # average tax unit size
    avg_size = table_values(data, 'ppl')
    # average UBI per person
    avg_ubi_person = table_values(data, 'ubi/person')
    # average payroll MTR
    avg_pmtr = table_values(data, 'pmtr')
    # average income tax MTR
    avg_imtr = table_values(data, 'imtr')
    # average combined MTR
    avg_cmtr = table_values(data, 'cmtr')
    # average tax change
    avg_tax_change = table_values(data, 'tax_change')
    # average UBI per tax unit
    avg_ubi = table_values(data, 'ubi')
    # average benefit change
    avg_ben = table_values(data, 'ben_change')
    # average benefit change - welfare adjusted
    avg_ben_welfare = table_values(data, 'ben_change_welfare')
    # average combined change
    avg_combined = table_values(data, 'combined_change')
    # average combined change - welfare adjusted
    avg_combined_welfare = table_values(data, 'combined_change_welfare')
    # tax units in a bin
    units = data.groupby('bins', as_index=False).sum()['s006']

    # Create DataFrame with all relevant information
    info = pd.DataFrame()
    if income_measure == 'e00200':
        income_str = 'Wage and Salary Floor (Thousands)'
    elif income_measure == 'c00100':
        income_str = 'AGI Floor (Thousands)'
    else:
        income_str = 'Income Floor (Thousands)'
    info[income_str] = (decile_floor / 1000).apply('{:,.0f}'.format)
    info['Tax Units (m)'] = (units / 1000000).apply('{:,.2f}'.format)
    info['Avg Tax Unit Size'] = avg_size.apply('{:,.2f}'.format)
    info['Avg UBI Per Person'] = avg_ubi_person
    info['Avg MTR - Payroll'] = (avg_pmtr * 100).apply('{:.1f}%'.format)
    info['Avg MTR - Individual Income'] = (avg_imtr *
                                           100).apply('{:.1f}%'.format)
    info['Avg MTR - Combined'] = (avg_cmtr * 100).apply('{:.1f}%'.format)
    info['Avg Tax Change'] = avg_tax_change
    info['Avg UBI Per Tax Unit'] = avg_ubi
    info['Avg Benefits Change'] = avg_ben
    info['Avg Benefits Change - Welfare Adj'] = avg_ben_welfare
    info['Avg Combined Change'] = avg_combined
    info['Avg Combined Change - Welfare Adj'] = avg_combined_welfare

    return info
```

```python
# create a table from the CPS data
cps_table_data = calc_cps_base.dataframe(['elderly_dependents', 'age_head',
                                          'age_spouse', 'e00200', 's006',
                                          'nu18', 'n1820', 'n21'])
cps_table_data['expanded_income'] = base_cps_data['expanded_income']
cps_table_data['ppl'] = cps_table_data['nu18'] + cps_table_data['n1820'] + cps_table_data['n21']
cps_table_data['ubi'] = reform_cps_data['ubi']
cps_table_data['ubi/person'] = cps_table_data['ubi'] / cps_table_data['ppl']
cps_tax_change = calc_cps_reform.array('combined') - calc_cps_base.array('combined')
cps_table_data['tax_change'] = cps_tax_change
cps_ben_change = (calc_cps_reform.array('benefit_cost_total') -
                  calc_cps_base.array('benefit_cost_total'))
cps_table_data['ben_change'] = cps_ben_change
cps_ben_change_welfare = (calc_cps_reform.array('benefit_value_total') -
                          calc_cps_base.array('benefit_value_total'))
cps_table_data['ben_change_welfare'] = cps_ben_change_welfare
cps_combined_change = (cps_table_data['ubi'] -
                       cps_table_data['tax_change'] +
                       cps_table_data['ben_change'])
cps_table_data['combined_change'] = cps_combined_change
cps_combined_change_welfare = (cps_table_data['ubi'] -
                               cps_table_data['tax_change'] +
                               cps_table_data['ben_change_welfare'])
cps_table_data['combined_change_welfare'] = cps_combined_change_welfare
cps_table_data['65_plus'] = np.where((cps_table_data['elderly_dependents'] == 1) |
                                     (cps_table_data['age_head'] >= 65) |
                                     (cps_table_data['age_spouse']), 1, 0)
cps_table_data['pmtr'] = cps_mtrs[0]
cps_table_data['imtr'] = cps_mtrs[1]
cps_table_data['cmtr'] = cps_mtrs[2]

# create a table from PUF data
puf_table_data = calc_puf_base.dataframe(['elderly_dependents', 'age_head',
                                          'age_spouse', 'e00200', 's006',
                                          'nu18', 'n1820', 'n21'])
puf_table_data['expanded_income'] = base_puf_data['expanded_income']
puf_table_data['ppl'] = puf_table_data['nu18'] + puf_table_data['n1820'] + puf_table_data['n21']
puf_table_data['ubi'] = reform_puf_data['ubi']
puf_table_data['ubi/person'] = puf_table_data['ubi'] / puf_table_data['ppl']
puf_tax_change = calc_puf_reform.array('combined') - calc_puf_base.array('combined')
puf_table_data['tax_change'] = puf_tax_change
puf_ben_change = (calc_puf_reform.array('benefit_cost_total') -
                  calc_puf_base.array('benefit_cost_total'))
puf_table_data['ben_change'] = puf_ben_change
puf_ben_change_welfare = (calc_puf_reform.array('benefit_value_total') -
                          calc_puf_base.array('benefit_value_total'))
puf_table_data['ben_change_welfare'] = puf_ben_change_welfare
puf_combined_change = (puf_table_data['ubi'] -
                       puf_table_data['tax_change'] +
                       puf_table_data['ben_change'])
puf_table_data['combined_change'] = puf_combined_change
puf_combined_change_welfare = (puf_table_data['ubi'] -
                               puf_table_data['tax_change'] +
                               puf_table_data['ben_change_welfare'])
puf_table_data['combined_change_welfare'] = puf_combined_change_welfare
puf_table_data['65_plus'] = np.where((puf_table_data['elderly_dependents'] == 1) |
                                     (puf_table_data['age_head'] >= 65) |
                                     (puf_table_data['age_spouse']), 1, 0)
puf_table_data['pmtr'] = puf_mtrs[0]
puf_table_data['imtr'] = puf_mtrs[1]
puf_table_data['cmtr'] = puf_mtrs[2]


income_bins = [0, 9999, 19999, 29999, 39999, 49999,
               74999, 99999, 199999, 1000000, 9e99]
```
